```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
```

```python
event_csv='traincasepowshell.csv'
event_column = ['File name','Index','Time','EntryType','Source','InstanceID','Message']
events=pd.read_csv(event_csv,sep=',',names=event_column,header=None)
events
```

```python
from sklearn.preprocessing import Imputer
events_X = events['EntryType']
events_Y = events['File name']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
level = events_X 
level_encoded = encoder.fit_transform(level)
level_encoded
```

```python

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
level_one_hot = encoder.fit_transform(level_encoded.reshape(-1,1))
events_X['EntryType'] = level_one_hot.toarray()[:557]

```

```python
X_train=level_one_hot
Y_train=events_Y

from sklearn.tree import DecisionTreeClassifier  
classifier = DecisionTreeClassifier()  
classifier.fit(X_train,Y_train)


#Split data into training and Testing data set. Ratio is 70:30
```

```python
dtest='testpowshell2.csv'
test_column= ['Index','Time','EntryType','Source','InstanceID','Message']
dtests=pd.read_csv(dtest,sep=',',names=test_column,header=None)
dtests

```

```python
from sklearn.preprocessing import Imputer
eventt_X = dtests['EntryType']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
levelt = eventt_X 
levelt_encodedt= encoder.fit(eventt_X)
level_encodedt = encoder.transform(levelt)
level_encodedt
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
leveltt= level_encodedt.reshape(-1,1)
level_one_hott = encoder.fit_transform(leveltt)
eventt_X['EntryType'] = level_one_hott.toarray()[:47]
```

```python
y_pred = classifier.predict(level_one_hott)
y_pred
```

```python
dtests='testpowshell2.csv'
test_column= ['Index','Time','EntryType','Source','InstanceID','Message']
dtestss=pd.read_csv(dtests,sep=',',names=test_column,header=None)
dtestss
```
